# Vanishing Gradient Problem

```python
#Importing mandatory libraries
import tensorflow as tf
import keras
from keras.layers import Dense
from keras.models import Sequential
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.model_selection import train_test_split
```

```python
import os
```

```python
os.getcwd()
```

```python
X,y=make_moons(n_samples=250,noise=0.05,random_state=42)
```

```python
X.shape
```

```python
y.shape
```

```python
X
```

```python
y
```

```python
plt.scatter(X[:,0],X[:,1], c=y,s=100)
plt.show()
```

```python
model=Sequential()
model.add(Dense(10,activation='sigmoid',input_dim=2))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(10,activation='sigmoid'))
model.add(Dense(1,activation='sigmoid')) # here we have taken 1 because it a classification problem.
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.get_weights()[0]
```

```python
old_weights=model.get_weights()[0]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=42)
model.fit(X_train,y_train,epochs=100)
```

```python
new_weight=model.get_weights()[0]
```

```python
old_weights
```

```python
new_weight
```

```python
percent_change=abs(100*(old_weights - new_weight)/old_weights)
```

```python
percent_change
```

```python
from mlxtend.plotting import plot_decision_regions
plot_decision_regions(X_train,y_train.astype('int'),clf=model,legend=2)
# If we observe below graph there is is  no separation boundary its not able to differentiate the classes(not able to learn- no decision boundary -encountered with vanishing gradient problem) as it is a classification problem(having two classes)
```
